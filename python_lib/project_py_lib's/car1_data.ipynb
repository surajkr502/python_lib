```python
df = pd.read_csv("D:/Ad_py_lib.csv/pandas_and_numpy/cars1.csv")
```

```python
numerical_cols = df.select_dtypes(include = ["float64","int64"]).columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for column in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x=column, bins=50, kde=False )
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
selected_features = ['cylinders','displacement','horsepower']
df['horsepower'] = pd.to_numeric(df['horsepower'])

for feature in selected_features:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x=feature, y='weight')
    plt.title(f'Scatter plot of {feature} vs weight')
    plt.xlabel(feature)
    plt.ylabel('weight')
    plt.show()

plt.figure(figsize=(12, 8))
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='magma', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

```
