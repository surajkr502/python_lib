```python
df=pd.read_csv("D:/Ad_py_lib.csv/project/california_housing_test.csv")
```

```python
numerical_cols=df.select_dtypes(include=['float64','int64']).columns
```

* This line collects all numeric column names (integers + floats) from your DataFrame and saves them in numerical_cols.
* df is your DataFrame (from Pandas).
* It contains rows (data records) and columns (features/variables).

```python
for columns in numerical_cols: 
    plt.figure(figsize=(10,6))
    sns.histplot(data=df,x=columns,bins=50,kde=True)
    plt.title(f'distribution of {columns}')
    plt.xlabel(columns)
    plt.ylabel('frequency')
    
    plt.show()
```

```python
selected_features = ['median_income', 'total_rooms', 'housing_median_age']
for feature in selected_features:
    plt.figure(figsize=(10,6))
    sns.scatterplot(data=df,x=feature, y='median_house_value')
    plt.title(f'Scatter plot of (featurer vs median_house_value')
    plt.xlabel(feature)
    plt.ylabel('median_house_value')
    plt.show()
plt.figure(figsize=(12,8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='magma', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
#Add a categorical column for visualizatÃ­on
df['house_value_category'] = pd.cut(df['median_house_value'],
            bins=[0, 100000, 200000,300000, 400000, 500000, float('inf')],
            labels=['<100k','100k-200k','200k-300k','300k-400k','400k-500k','>500k'])

# Visualize the distribution of house value categories
plt.figure(figsize=(10,6))
sns.countplot(data=df,x='house_value_category',order=df["house_value_category"].value_counts().index,palette='viridis')
plt.title('Distribution of Median House Value Categories')
plt.xlabel('Median House Value Category')
plt.ylabel('Count')
plt.show()
# visualize the relation between median income and median house value,colored by house value category 
plt.figure(figsize=(12,8))
sns.scatterplot(data=df,x='median_income',y='median_house_value',hue='house_value_category',
                size='population', alpha=0.6,sizes=(20,200),palette='viridis')
```
