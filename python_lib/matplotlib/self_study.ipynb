```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)       # Line plot
plt.xlabel("X-axis") # Label
plt.ylabel("Y-axis")
plt.title("Basic Line Plot")
plt.show()
```
