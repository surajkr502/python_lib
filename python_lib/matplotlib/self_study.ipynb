```python
import matplotlib.pyplot as plt 
# plt.plot(x,y,color='color name',linestyle='line style',linewidth=value,marker='marker symbol',label='lebel name')
# plt.xlabel('text')
# plt.xlabel('text')
# plt.xlim(start,end)
x=[1,2,3,4,5,6]
y=[18000,25000,60000,50000,20000,35000]
plt.plot(x,y,color='blue',linestyle='--',linewidth=1,marker='o',label='2025 sales data')
plt.xlabel('Months')
plt.ylabel('Sales Per Months')
plt.title('Monthly Sale Data Report')
plt.legend(loc='upper right',fontsize=10)
plt.grid(color='gray',linestyle=':',linewidth=0.5)
plt.xlim(1,7)
plt.ylim(0,65000)
plt.xticks([1,2,3,4,5,6],['M1','M2','M3','M4','M5','M6'])
plt.show()
```
