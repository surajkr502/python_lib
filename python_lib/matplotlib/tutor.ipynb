```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
```

### Byvariant Data(2D variant)

```python
plt.plot(data['Year'],data['Virat'])
plt.xlabel('years')
plt.ylabel('Score')
```

```python
plt.plot(data['Virat'],data['Rohit'])
```

```python
plt.plot(data['Year'],data['Dhoni'],color='red',label='Dhoni',linestyle='dotted',marker='D',linewidth='1.5')
plt.plot(data['Year'],data['Rohit'],color='green',label='Rohit',linestyle='solid',marker='o')
plt.plot(data['Year'],data['Virat'],color='#368588',label='Virat',linestyle='dashed',marker='s') #color code
plt.title("Score comparison")
plt.xlabel('Years')
plt.ylabel('Score')
plt.legend(loc='best') #by default take best
plt.grid()
plt.show()
```
