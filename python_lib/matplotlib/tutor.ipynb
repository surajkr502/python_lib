```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv("cricket.csv")
```

```python
data
```

```python
plt.plot(data['Year'],data['Virat'])
plt.xlabel('years')
plt.ylabel('Score')
```

```python
plt.plot(data['Virat'],data['Rohit'])
```

```python
plt.plot(data['Year'],data['Dhoni'],color='red',label='Dhoni')
plt.plot(data['Year'],data['Rohit'],color='green',label='Rohit')
plt.plot(data['Year'],data['Virat'],color='#368588',label='Virat') #color code
plt.title("Score comparison")
plt.xlabel('Years')
plt.ylabel('Score')
plt.legend(loc='best') #by default take best
```
