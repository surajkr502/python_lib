### Broadcasting

```python
# j=[45,64,76,46,43,56,33,56,76]
# for i in j:
#     i=i+10
#     print(i)
```

```python
j=[23,53,36,32,63,45,78,100]
l=len(j)
for i in range(l):
    j[i]=j[i]+10  #Broadcasting
print(j)
```

```python
sr=pd.Series(j)
print(sr)
sr.plot(kind='barh')
```
