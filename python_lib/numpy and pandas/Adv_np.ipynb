## Numpy array vs Python list

#### Speed

```python
# speed of list 
import time
# start=time.time() 
a=[i for i in range(10000000)]
b=[i for i in range(10000000,20000000)] 
c=[]
start=time.time() 
for i in range(len(a)):
    c.append(a[i]+b[i])
print(time.time()-start)
 
```

```python
# speed of numpy  
import numpy as np 
a=np.arange(10000000)
b=np.arange(10000000,20000000)
start=time.time() 
c=a+b 
print(time.time()-start)
```

#### Memory

```python
a=[i for i in range(10000000)]
import sys
sys.getsizeof(a)
```

```python
a=np.arange(10000000,dtype=np.int32)
sys.getsizeof(a)
```

## Advanced Indexing

### Fancy Indexing

```python
a=np.arange(24).reshape(6,4)
a
```

```python
a[[0,2,3,5]]
```

```python
a[:,[0,2,3]]
```

```python
a[[1,3,4],[0,2,3]]
```

### Boolean Indexing

```python
a=np.random.randint(1,100,24).reshape(6,4)
a
```

```python
# Find all no. greater than 50. 
a[a>50]
```

```python
# Find all even No. 
a[a%2==0]
```

```python
# Greater than 50 or even no. 
a[(a%2==0) & (a>50)] #use bitewise operater(&) with boolean value instead of logical operater(and). 
```

```python
# Not divisible by 7.
a[a%7!=0]
a[~(a%7==0)]      #both are same
```

## Broadcasting 

```python
# same shape 
a=np.random.randint(1,7,6).reshape(2,3)
b=np.random.randint(6,13,6).reshape(2,3)
print(a)
print(b)
c=a+b
```

```python
# different shape 
a=np.arange(6).reshape(2,3)
b=np.arange(6,9).reshape(1,3)
print(a)
print(b)
c=a+b
```

```python
a=np.arange(6).reshape(2,3)
b=np.arange(3)
print(a)
print(b)
c=a+b 
print(c)
```

```python
a=np.arange(12).reshape(4,3)
b=np.arange(3)
print(a)
print(b)
c=a+b
print(c)
```

```python
a=np.arange(3).reshape(3,1)
b=np.arange(3,6).reshape(1,3)
print(a+b)
```

```python
a=np.arange(3).reshape(1,3)
b=np.arange(3,6).reshape(3,1)
print(a+b)
```

```python
a=np.array([1])
b=np.arange(4).reshape(2,2)
print(a+b)
```

 ## Working with Mathematical formulas

```python
# sigmoid
def sigmoid(array): 
    return 1/(1+np.exp(-(array)))
a=np.arange(10)
sigmoid(a)
```

```python
# Mean Square Error
actual=np.random.randint(1,50,25)
predicted=np.random.randint(1,50,25)
def MSE(actual,predicted): 
    return np.mean((actual-predicted)**2)
print(MSE(actual,predicted))
```
