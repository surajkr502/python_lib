## Series DataFrame. 
* 1D array or column is called Series DataFrame.

```python
Ar=np.array([10,20,30,40])
Ar.dtype
Ar.ndim
type(Ar)
```

```python
At=[1,2,3,4,5,6,7]
type(At)
# pd.Series(At) #series generated by list.
pd.Series(Ar)  #series generated by numpy array.
```

```python
dic={
    'ram':20,'mohan':56
}
pd.Series(dic)
```

```python
t=(1,2,3,4,56,76)
pd.Series(t)
```

```python
A=[34,53,64,57,59,60,38,73,94,20]
A
B=['cpp','python','java','c#']
s=pd.Series(A)
s
```

```python
s1=pd.Series(B,name='cybrom') 
s1
```

```python
print(s.size)
print(s1.size)
print(s.dtype)
print(s1.dtype)
print(s.is_unique)
print(s.name)
print(s1.name)
print(s.index)
print(s.values)
```

```python
t=(32,46,32,64,35,74,20,46)
st=pd.Series(t)
print(st.values)
```

```python
print(st.value_counts())
```

```python
A=[34,56,34,74,85,16,63,20,80,30,25,843,32,687,123,456]
s=pd.Series(A)
s
print(s.sample(3))
```

```python
print(s.value_counts())
```

```python
print(s.sort_values()) 
```

```python
print(s.sort_index())
```

```python
print(s.sort_values().head(1).values[0])
```

### customized index

```python
score=[54,60,63,27,48,95,63]
Name=['a','b','c','d','e','f','g']
s=pd.Series(score,index=Name)
s
```

```python
A=[23,53,56,75,34,56,75,43,64,75,35,45,96,33,64,24]
s=pd.Series(A)
```

```python
D=np.array([2,4,np.nan,5,6,np.nan])
print(D)
SD=pd.Series(D)
print(SD.isnull())
print(SD.hasnans)
print(SD.notnull())
print(SD.unique()) 
print(SD.nunique())
```

```python
print(len(dir(SD)))
```
