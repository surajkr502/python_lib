### Stacking
* Horizontal Stacking
* Vertical Stacking

```python
# Horizontal Stacking 
h1=np.arange(12).reshape(3,4)
h2=np.arange(12,24).reshape(3,4)
print(h1)
print(h2)
np.hstack((h1,h2))
```

```python
print(h1)
print(h2)
np.vstack((h1,h2))
```

 ### Splitting
 * Horizontal Splitting
 * Vertical Splitting

```python
# Horizontal Splitting 
print(h1)
np.hsplit(h1,2) #must be equal division
```

```python
# Vertical Stacking 
print(h1)
# np.vsplit(h1,2) #must be equal division 
np.vsplit(h1,3)
```
